```python
from app_store_scraper import AppStore
from pprint import pprint

ognomy = AppStore(country="us", app_name="ognomy")
ognomy.review(how_many=50)

pprint(ognomy.reviews)
pprint(ognomy.reviews_count)
```

```python
final_data_list=[]

for i, review in enumerate(ognomy.reviews):
  all_detail_list=[]
  all_detail_list=[review['date'], review['isEdited'], review['rating'], review['review'],\
                   review['title'], review['userName']]

  final_data_list.append(all_detail_list)
```

```python
import pandas as pd

final_df= pd.DataFrame(final_data_list, columns =["Comment_time","review_is_edited?","reviewRating", "User_Comment", "reviewTitle"\
                                              ,"userName"])
```
